```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv("prediction.csv")

# Extract features (X) and target variable (y)
X = df[['area']]  # Use double square brackets to select a DataFrame with a single column
y = df['prices']  # Assuming 'prices' is the correct column name

# Visualize the data
plt.scatter(X, y, color="green", marker="+")
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Price vs Area")
plt.show()
```

```python
reg = LinearRegression()
reg.fit(X, y)
```

```python
reg.predict([[3000]])
```
